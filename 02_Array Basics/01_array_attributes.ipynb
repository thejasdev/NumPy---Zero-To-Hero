1. Array Attributes

```python
import numpy as np
```

```python
# After Creating the Numpy Arrays , Attributes helps to understand the structure of Array
# Determining the ndim, shape, size, dtype, nbytes, 
# Following Example on array attributes
# 1D Array
arr = np.array([1, 2, 3, 4, 5])
print(arr)
print("Dimension :",arr.ndim) # ndim --> Number of Dimensions
print("Shape :",arr.shape) #shape --> Gives the shape of the array as tuple
print("Size :", arr.size) #size --> Gives the total number of elements present in array
print("DataType :",arr.dtype) # dtype --> Classifies the datatype of array
print("Number of Bytes :",arr.nbytes)
```
